Ainda não definimos rigorosamente as superfícies, mas você provavelmente tem
uma ideia intuitiva do que é uma superfície. É como um pedaço de papel, mas não

## Parametrizar uma superficie

Muitas vezes nossas superfícies são (um pedaço de) o gráfico de uma função
$z = f (x, y)$, como neste exemplo para $z=x^2 + y^2$:

```python
av.plot3d(x**2+y**2, (x,-1,1),(y,-1,1))
```

O Plotly pode desenhar ótimos gráficos como este, mas o `av.plot3d` tem suas
limitações. Por exemplo, se quiséssemos plotar o gráfico de $x = y^2 + z^2$ ? Você
deve saber, olhando para esta equação, que este é um paraboloide, assim como a
nossa imagem acima, exceto que este se abre na direção do eixo x positivo.
Suponha que tentássemos traçar isso com `av.plot3d` :

```python
av.plot3d(z**2+y**2, (y,-1,1),(z,-1,1))
```

Nós temos exatamente a mesma imagem de antes! Mas olhando com calma
percebemos que o paraboloide abre para cima, o que é incorreto! A razão é que
`av.plot3d` espera uma função de duas variáveis e interpreta os valores da função
como a altura. Normalmente, damos a ele uma função de $x$ e $y$, e assim o eixo $z$ representa a altura, isso se encaixa em nossa visão do mundo.
Como podemos obter um gráfico correto de $x = y^2 + z^2$ do Plotly? Uma possibilidade é usar as equações paramétricas. A gente já viu como fazer uma “parametrização trivial” de uma curva plana (2D) tipo $y = g(x)$. a parametrização fica $\vec r (x) = (x, g(x))$. Na mesma forma podemos parametrizar a superfície acima usando $y$ , $z$ como os parâmetros. Defina

$$
\vec r = (y^2 + z^2)\hat i + y \hat j + z\hat k = \begin{cases}
x(y,z) &=y^2 + z^2 \\
y(y,z)&=y \\
z(y,z) &= z
\end{cases}
$$

Agora podemos traçar o gráfico de $\vec r$ usando `av.plot3d_parametric_surface`. Esse metodo utiliza a função `add_surface` do Plotly, ['3D Surface Plots in Python'](https://plotly.com/python/3d-surface-plots/)

```python
def surf(y,z):
    return (y**2+z**2)*R.i + y*R.j + z*R.k
```

```python
av.plot3d_parametric_surface(surf(y,z), (y,-1,1),(z,-1,1))
```

E você pode ver que esta é a abertura da paraboloide é na direção positiva do eixo $x$, conforme desejado.
Observe que estamos usando dois parâmetros agora, o que é diferente de quando

plotamos curvas. Isso faz sentido, se você pensar sobre isso. Curvas são como uma linha, um objeto unidimensional, então elas requerem um parâmetro. As superfícies são criaturas bidimensionais e, portanto, exigem dois parâmetros.

## Mais Exemplos

Vamos ver mais exemplos de como parametrizar superfícies. A gente já sabe que as regiões circulares são muito importantes, e há uma técnica que você deve aprender. Muitas de nossas superfícies serão relativamente fáceis de descrever usando
coordenadas cilíndricas $(\rho, \theta, z)$, que foram cobertas no calculo2. Porem, uma parametrização tem que ser em coordenadas retangulares $( x, y, z )$ , então não podemos simplesmente descrever a superfície em coordenadas cilíndricas e dizer que terminamos. Mas, se descrevemos a superfície em coordenadas cilíndricas, podemos usar as seguintes fórmulas para converter a nossa descrição em coordenadas retangulares:
$$
\begin{cases}
x =& \rho \cos(\theta) \\
y =& \rho \sin\theta) \\
z =& z
\end{cases}
$$



### Exemplo 1: Discos de Raio $R$ no Plano $z = h$

Em coordenadas cilíndricas, tal disco é descrito por $(\rho, \theta, h)$ onde $0\le \rho \le R$, $\; 0\le \theta \le 2\pi$, e $h$ é algum número constante. Convertendo para coordenadas retangulares, temos $\vec r (\rho,\theta)=(\rho \cos(\theta), \rho\sin(\theta), h)$, onde $\rho$ e $\theta$ tem os mesmos limites ($\rho$ e $\theta$ são nossos parâmetros agora.) Por exemplo, se $R = 1$ e $h = 1$


```python
def disc(rho, theta,h=1):
    return rho*sp.cos(theta)*R.i + rho*sp.sin(theta)*R.j + h*R.k
```

```python
disc(rho, theta)
```

```python
fig = av.plot3d_parametric_surface(disc(rho, theta), 
                             (rho,0,1),(theta,0,2*np.pi))

fig
```

Ou, com outros valores de $h$:

```python
av.plot3d_parametric_surface(disc(rho, theta,h=2), 
                             (rho,0,1),(theta,0,2*np.pi), fig=fig)

av.plot3d_parametric_surface(disc(rho, theta, h=3), 
                             (rho,0,1),(theta,0,2*np.pi),fig=fig)
```

### Exemplo 2: um anel

Na verdade, não usaremos muito isso, mas é um exemplo interessante a ser destacado.Suponha que nós parametrizamos o disco no plano $z = 1$, porem o raio vai variar entre 2 e 4:

```python
fig = av.plot3d_parametric_surface(disc(rho, theta), 
                             (rho,2,4),(theta,0,2*np.pi))

fig
```

### Exemplo 3: cone

A parametrização do cone é dada por $\vec r (\rho,\theta)=(\rho \cos(\theta), \rho \sin(\theta), \rho)$. Vamos plotar um cone de raio 4

```python
def cone(rho, theta,h=1):
    return rho*sp.cos(theta)*R.i + rho*sp.sin(theta)*R.j + rho*R.k
```

```python
fig = av.plot3d_parametric_surface(cone(rho, theta), 
                             (rho,0,4),(theta,0,2*np.pi))

fig
```

```python
fig = av.plot3d_parametric_surface(cone(rho, theta), 
                             (rho,2,4),(theta,0,2*np.pi))

fig
```
